```python
import pyspark
from pyspark.sql import SparkSession
from pyspark.sql.types import *
```

```python
spark=SparkSession.builder.appName('Reading csv file').getOrCreate()
spark
```

```python
help(spark.read)
```

```python
df_pyspark=spark.read.csv(path='test.csv')# we can even directly pass the path no need to give path=
df_pyspark.show()
```

```python
df_pyspark=spark.read.csv(path='test.csv',header=True)# we can even directly pass the path no need to give path=
df_pyspark.show()
```

```python
df_pyspark=spark.read.csv('test.csv',header=True,inferSchema=True)# we can even directly pass the path no need to give path=
df_pyspark.show()
```

```python
df_pyspark=spark.read.option('header','True').csv(path='test.csv')# we can even directly pass the path no need to give path=
df_pyspark.show()
df_pyspark.printSchema()
```

```python
df_pyspark=spark.read.format('csv').load(path='test.csv')# we can even directly pass the path no need to give path=
df_pyspark.show()
```

```python
df_pyspark=spark.read.format('csv').option('header','True').load(path='test.csv',inferSchema=True)# we can even directly pass the path no need to give path=
df_pyspark.show()
df_pyspark.printSchema()
```

```python
df_pyspark=spark.read.option('header','True').format('csv').load(path='test.csv')# we can even directly pass the path no need to give path=
df_pyspark.show()
```

```python
df_pyspark=spark.read.csv(['test.csv','test1.csv'],inferSchema=True,header=True)
df_pyspark.show()
df_pyspark.printSchema()
```
